```python
import numpy as np
import glob

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer

import tensorflow as tf

from keras.models import Sequential
from keras.layers.core import Dense, Activation, Flatten
```

```python
#Load data
label_binarizer = LabelBinarizer()

X = []
Y = []

label = 0
for i in glob.glob('Features/*'):
    for j in glob.glob(i+'\\*'): # \\ for windows
        temp_data = np.loadtxt(j).reshape(-1)
        X.append(temp_data)
        Y.append(label)
    label+=1

X = np.array(X)
Y = np.array(Y)

X_ = (X - np.mean(X)) / np.std(X)
Y_ = label_binarizer.fit_transform(Y)

X_train, X_test, Y_train, Y_test = train_test_split(X_, Y_, test_size=0.3)
```

```python
sz_inp = X_.shape[1]
sz_out = Y_.shape[1]

model = Sequential()
model.add(Dense(sz_out, input_shape=(sz_inp, )))
model.add(Activation('softmax'))
model.summary()
```

```python
model.compile(optimizer = 'sgd', metrics=['accuracy'], loss='categorical_crossentropy')
history = model.fit(X_train, Y_train, epochs=1000, batch_size=1024*10)
```

```python
model.evaluate(X_test, Y_test)
```
